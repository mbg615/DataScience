## 6 - Droping and Renaming Columns

```python
df_realestatecopy = df_realestate.copy() # Make copy of the data
```

```python
df_realestatecopy.drop(columns = ['Zoning Class','Lot Shape','Lot Config',"Land Slope",'Bldg Type','House Style','Roof Style','Roof Material','Exterior Primary','Masonry/Veneer',
                              'Exterior Qual','Exterior Cond','Foundation','Basement Height','Basement Cond','Basement Exposure','Basement Finish','Heating Qual','CentralAir','Electrical',
                             'Functionality','Fireplce Qual','Garage Type','Garage Qual','Garage Cond','Paved Drive','Pool Qual','Fence','Sale Type','Year Remod Add'])
```

```python
df_realestatecopy = df_realestatecopy.rename(columns = {'Type':'Dwelling Type','QvQual':'Overall Quality','Nbhd':'Neighborhood','Built':'Year Built'}) # Rename columns
```

## 7 - Grouping Data

## 7a - Group by Neighborhood and replace values

```python
df_realestatecopy.groupby('Neighborhood')['Neighborhood'].count() # Group by Neighborhood
```

```python
df_realestatecopy = df_realestatecopy.replace({'Bloomington Hts':'Bloomington Heights'}) # Replace Bloomington Hts with Bloomington Heights
df_realestatecopy.groupby('Neighborhood')['Neighborhood'].count() # Group by Neighborhood
```

## 7b-Group based on 'Dwelling Type' and replace values

```python
df_realestatecopy.groupby(['Dwelling Type'])['Dwelling Type'].count() # Group by Dwelling Type
```

```python
df_realestatecopy = df_realestatecopy.replace({'1-STORY PUD':'1 STORY PUD'}) # Replace 1-STORY PUD with 1 STORY PUD
df_realestatecopy.groupby(['Dwelling Type'])['Dwelling Type'].count()
```

```python
df_realestatecopy.groupby(['Neighborhood','Dwelling Type'])['Sale Price'].median()
```

## 8 - Summarizing and Filtering data

## 8a - Pivot Neighborhood and Land Contour

```python
df_re_pivot = df_realestatecopy.pivot_table(index='Neighborhood',columns='Land Contour',values='Sale Price') # pivot the data file with 'Neighborhood index, 'Land Contour' columns, 'Sale Price' values
df_re_pivot
```
